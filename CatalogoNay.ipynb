*   **Uniforme**

---



---





$\mathcal{U}(a,b) \qquad a<b$

$f(x)=$  
\begin{cases}
    \frac{1}{b-a} & \text{para } a \leq x \leq b \\
    0 & \text{en otro caso}
\end{cases}
image.png

$F(x)=$
\begin{cases}
    0 & \text{para } x < a \\

image.png

*PSEUDOCODIGO*

FUNCIN unif(a, b, n):

    Crear lista vaca

    PARA i desde 1 hasta n:

        Generar un nmero aleatorio R uniformemente distribuido entre 0 y 1

        Calcular valor transformado: resultado = (R * b) + a * (1 - R)

        Agregar resultado a la lista

    FIN PARA

    RETORNAR

FIN FUNCIN

```python
import random
import matplotlib.pyplot as plt
def unif(a,b,n ):
  lista=[]
  for i in range(n):
    R=random.random()
    lista.append((R*b)+a*(1-R))
  return lista

exper = unif(1,6,10000)

plt.hist(exper, bins=1000,color="pink")
plt.show()
```



*   **Triangular**


---



---





$\mathcal{Tri(a,b, c)}$

$f(x) =$ \begin{cases}
    \frac{2(x-a)}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
    \frac{2(b-x)}{(b-a)(b-c)} & \text{para } c \leq x \leq b \\
    0 & \text{en otro caso}
  \end{cases}
image.png
  $F(x)=$ \begin{cases}
    0 & \text{para } x < a \\
    \frac{(x-a)^2}{(b-a)(c-a)} & \text{para } a \leq x \leq c \\
    1 - \frac{(b-x)^2}{(b-a)(b-c)} & \text{para } c \leq x \leq b \\
    1 & \text{para } x > b
  \end{cases}
image.png
  $E[X]=\frac{a+b+c}{3}$

  $V[X]=\frac{a^2+b^2+c^2-ab-ac-bc}{18}$

  *PSEUDOCODIGO*

  FUNCIN tri(a, b, c, n):

    Crear lista vaca

    PARA i desde 1 hasta n:

        Generar un nmero aleatorio R uniformemente distribuido entre 0 y 1

        Calcular primer posible valor de x:
            xi = [ R * (c - a) * (b - a) ] + a

        SI xi est entre a y b:
            Agregar xi a la lista
        SINO:
            Calcular segundo posible valor de x:
                xi = c - [ (1 - R) * (c - a) * (c - b) ]
            Agregar xi a la lista
    FIN PARA
    RETORNAR lista
FIN FUNCIN

```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math

def tri(a,b,c,n ):
    lista=[]#lista vaca para almacenar las x calculadas
    for i in range(n):#iniciamos ciclo para repetir el proceso n veces
       R = random.random()#generamos el nmero aleatorio
       xi = np.sqrt(R*(c-a)*(b-a))+a #operacin del primer valor de x
       if a <= xi <= b: # definimos el condicional
          lista.append(xi) # si cumple las condiciones se almacena en la lista
       else:# sino las cumple pasamos a clcular el otro valor de x
         xi = c-np.sqrt((1-R )*(c-a )*(c-b ))
         lista.append(xi)
    return lista #regresa la lista de los n valores de x calculados

#definimos los parmetros iniciales
a=2
b=4
c=6
#las veces que se repetir el experimento
n=10000
#asignamos una variable al experimento
h =tri(a,b,c,n)
#graficamos nuestra funcin de distribucin Triangular
plt.hist(h, bins=1000, color="pink")
plt.show()
```

FUNCIN exponencial(lambda, n):



*  **M-Erlang**


---



---




$\mathcal{Erlang(\lambda, n)} \qquad \lambda >0 \qquad n \in \mathbb{Z}^+ \qquad x \in [0,\infty)$
  
$f(x) =$ \begin{cases}
    \frac{\lambda (\lambda x)^{n-1} e^{-\lambda x}}{(n-1)!} & \text{para } x \geq 0 \\
    0 & \text{para } x < 0
  \end{cases}
  image.png

$F(x) = 1 - \sum_{k=0}^{n-1} \frac{(\lambda x)^k e^{-\lambda x}}{k!}$

image.png
  $E[X] = \frac{n}{\lambda}$


  $\sigma^2 = \frac{n}{\lambda^2}$

  *PSEUDOCDIGO*

  FUNCIN erlang(m, lambda, n):

    Crear lista vaca

    Calcular mu = m / lambda  # media de la distribucin

    PARA i desde 1 hasta n:

        Crear lista U con m nmeros aleatorios uniformes entre 0 y 1

        Calcular producto = producto de todos los valores en U

        Calcular valor: xi = -(mu / m) * ln(producto)

        Agregar xi a la lista

    FIN PARA

    RETORNAR lista

FIN FUNCIN

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

def erlang(m, lanbda,n):
  lista = []
  mu = m / lanbda #Media
  for i in range(n):
    U = [random() for _ in range(m)]
    producto = np.prod(U)
    lista.append(-(mu / m) * np.log(producto))
  return lista

exp = erlang(4, 5,10000)
plt.hist(exp, bins=1000,color="pink")
plt.show()
```



*   **Gamma**


---



---




$\mathcal{Gamma(, )} \qquad \alpha, \beta >0 \qquad x \in [0,\infty )$

Tomando $m=\alpha \qquad =\frac{1}{}$

$\frac{\lambda^m x^{m-1} e^{-\lambda x}}{(m-1)!}  \frac{\lambda^m x^{m-1} e^{-\lambda x}}{\Gamma(m)}$

$f(x) =$ \begin{cases}
    \frac{\beta^{-\alpha} x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

con $\Gamma(k+1)=k!$

$F(x)=$ \begin{cases}
    1- e^{-\frac{x}{\beta}}\sum_{i=0}^{\alpha -1}\frac{(\frac{x}{\beta})^j}{j!} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}
  $E[X] = \alpha \beta$


  $V[X] = \alpha \beta^2$

*PSEUDOCDIGO*

FUNCIN gamma(alpha, n):

    Crear lista vaca

    Definir e = nmero de Euler (aprox. 2.718)

    Calcular constante b = (e + alpha) / e

    REPETIR n veces:

        Generar nmero aleatorio U1 entre 0 y 1

        Calcular P = b * U1

        SI P > 1:
            Calcular Y = -ln((b - P) / alpha)

            Generar otro nmero aleatorio U2 entre 0 y 1

            SI U2 <= Y^(alpha - 1):

                Agregar Y a la lista

        SINO:
            Calcular Y = P^(1 / alpha)

            Generar otro nmero aleatorio U2 entre 0 y 1

            SI U2 <= e^(-Y):

                Agregar Y a la lista


    FIN REPETIR

    RETORNAR lista
FIN FUNCIN

```python
import numpy as np
import matplotlib.pyplot as plt
import random
#0<alpha<1 y beta=1
def gamma_1(al,n):
    lista=[]
    e = np.e
    b = (e + al) / e
    for _ in range(n):
        U1=random.random()
        P=b*U1
        if P>1:
            Y=-np.log((b-P)/al)
            U2=random.random()
            if U2<=Y**(al-1):
                lista.append(Y)
        else:
            Y=P**(1/al)
            U2=random.random()
            if U2<=e**(-Y):
                lista.append(Y)
    return lista
G=gamma_1(1/2 ,10000)
print("Media aproximada:",np.mean(G))
print("Varianza aproximada: ",np.var(G))
plt.hist(G, bins=1000,color="pink")
plt.show()
```

*PSEUDOCODIGO*

FUNCIN gamma2(alpha, n):

    SI alpha  1:

        Lanzar error: "Vlido solo para alpha > 1"

    Calcular constantes:

        be  1 / sqrt(2 * alpha - 1)

        b  alpha - ln(4)

        q  alpha + 1 / be

        theta  4.5

        d  1 + ln(theta)

    Crear lista vaca x para almacenar los valores generados

    REPETIR n veces:

        Generar U1  nmero aleatorio uniforme entre 0 y 1

        Generar U2  nmero aleatorio uniforme entre 0 y 1

        Calcular V  be * ln(U1 / (1 - U1))

        Calcular Y  alpha * exp(V)

        Calcular Z  U1 * U2

        Calcular W  b + q * V - Y


        SI W + d - theta * Z  0:

            Agregar Y a la lista
        SINO SI W  ln(Z):

            Agregar Y a la lista


    FIN REPETIR

    RETORNAR lista

FIN FUNCIN


```python
import numpy as np
import matplotlib.pyplot as plt
import random

def gamma2(alpha,n):
  if (0 < alpha < 1):
    raise ValueError("Vlido para alpha > 1.")

  beta = 1 / np.sqrt(2 * alpha - 1)
  b = alpha - np.log(4)
  q = alpha + 1 / beta
  theta = 4.5
  d = 1 + np.log(theta)

  lista = []
  for _ in range(n):
    U1 = random.random()
    U2 = random.random()
    V = beta * np.log(U1 / (1 - U1))
    Y = alpha * np.exp(V)
    Z = U1 ** 2 * U2
    W = b + q * V - Y
    if W + d - theta * Z >= 0:
      lista.append(Y)
    elif W >= np.log(Z):
      lista.append(Y)
  return lista

G2 = gamma2(2,10000)
plt.hist(G2, bins=1000,color="pink")
print("Media aproximada:",np.mean(G2))
print("Varianza aproximada: ",np.var(G2))
```

$\mathcal{We(\alpha, \beta)} \qquad \alpha, \beta >0 \qquad x \in (0, \infty)$

$f(x)=$\begin{cases}
    \alpha \beta^{-\alpha}x{\alpha - 1}e^{-(x/\beta)^\alpha} & \text{para } x > 0 \\
    0 & \text{otro caso}
  \end{cases}

$= \lambda\alpha(\lambda x)^{\alpha-1}e^{(\lambda x)\alpha}$

$F(x)=$\begin{cases}

    0 & \text{otro caso}
  \end{cases}
$= 1-e^{-(\lambda x)^\alpha}$
image.png

$E[X]=\frac{\beta}{\alpha}\Gamma(\frac{1}{\alpha})=\frac{1}{\lambda}\Gamma(1+\frac{1}{\alpha})$

$V[X]=\frac{\beta^2}{\alpha}[2\Gamma(\frac{2}{\alpha})-\frac{1}{\alpha}[\Gamma(\frac{1}{\alpha})]^2]= \frac{1}{\lambda^2}[\Gamma(1+\frac{2}{\alpha})-\Gamma^2(1+\frac{1}{\alpha})]$

*PSEUDOCDIGO*

FUNCIN weibull(beta, alpha, n):

    Crear lista vaca

    REPETIR n veces:

        Generar nmero aleatorio U entre 0 y 1 (uniforme)

        Calcular xi = beta * ( -ln(U) )^(1 / alpha)

        Agregar xi a la lista

    FIN REPETIR

    RETORNAR lista x
    
FIN FUNCIN

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def weibull(beta,alpha, n):
  lista = []
  for _ in range(n):
    U = random.random()
    lista.append(beta*(-np.log(U))**(1/alpha))
  return lista

w = weibull(1,3,10000)
plt.hist(w, bins=1000,color="pink")
print("Media aproximada:",np.mean(w))
print("Varianza aproximada: ",np.var(w))
```

*PSEUDOCDIGO BOX-MULLER*
Algoritmo BoxMuller
Entrada:
    n: pares a genera
Salida:
    lista

Inicio
    # Inicializar lista vaca
    
    # Generar n pares de variables normales
    Para i desde 1 hasta n hacer:

        # Paso 1: Generar dos nmeros aleatorios uniformes U1 y U2 en (0, 1)

        R  RaizCuadrada(-2 * Ln(U1))      

        Theta  2 *  * U2                  

        X  R * Coseno(Theta)

        Y  R * Seno(Theta)
        
        # Paso 4: Almacenar ambos valores

        lista.Agregar(X)

    Fin Para

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def BoxM(n):
    lista=[]#lista para almacenar las desviaciones
    for _ in range(n):#ciclo para generar los nmeros aleatorios
        u1=rd.random()
        u2=rd.random()
        #Parmetros
        r = np.sqrt(-2*np.log(u1))
        theta=2*np.pi*u2
        x=r*np.cos(theta)
        y=r*np.sin(theta)
        lista.append(x)#almacena la desviacin x
        lista.append(y)#almacena la desviacin y
    return lista
BoxM = BoxM(10000)
print("Media aprox:",np.mean(BoxM))
print("Varianza aprox: ",np.var(BoxM))
print("Desviacin estandar aprox: ",np.std(BoxM))
plt.hist(BoxM, bins=1000,color="pink")
plt.show()
```

$f(x) =$\begin{cases}

*PSEUDOCDIGO*

Entrada:

    mup: Media deseada de la distribucin log-normal

    sigp: Desviacin estndar deseada de la distribucin log-normal

    n: Nmero de muestras a generar .
Salida:

    Lista de n valores generados con distribucin log-normal.

Inicio

    mu  ln(mup^2 / sqrt(mup^2 + sigp^2))

    sigma  sqrt(ln(1 + sigp^2 / mup^2))
    
    # Inicializar lista vaca para almacenar los valores generados
    
    lista  []
    
    # Generar n muestras

    Para i desde 1 hasta n hacer:

        # Generar Y ~ N(mu, sigma) usando la normal estndar

        lista.Agregar(valor_generado)

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def log_normal(mup,sigp,n):
  mu=np.log(mup**2/np.sqrt(mup**2+sigp**2))
  sigma=np.sqrt(np.log(1+sigp**2/mup**2))
  lista=[]
  for _ in range(n):#ciclo para generar los nmeros aleatorios
    Y=np.random.normal(mu,sigma)
    lista.append(np.exp(Y))#almacen de la desviacin generada
  return lista

Ln=log_normal(1,1,1000)
plt.hist(Ln, bins=100,color="pink")
print("Media aprox:",np.mean(Ln))
print("Varianza aprox: ",np.var(Ln))
print("Desviacin estandar aprox: ",np.std(Ln))
```

$\mathcal{\beta(\alpha,\beta)} \qquad \alpha,\beta>0$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def beta(al,be,n):
  lista=[] #lista para almacenar las desviaciones
  for _ in range(n):#ciclo para generar los nmeros aleatorios
    Y1=np.random.gamma(al,1)
    Y2=np.random.gamma(be,1)
    lista.append(Y1/(Y1+Y2))#almacen de las X generadas
  return lista

k=beta(0.5,0.5,1000)
plt.hist(k, bins=100,color="pink")
print("Media aprox:",np.mean(k))
print("Varianza aprox: ",np.var(k))
print("Desviacin estandar aprox: ",np.std(k))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def pearson_v(al,be,n):
  lista=[]
  for _ in range(n):
    Y=np.random.gamma(al,1/be)
    lista.append(1/Y)
  return lista

P5 = pearson_v(2,1,10000)

plt.hist(P5, bins=1000,range=(0,5),color="pink")
print("Media aprox:",np.mean(P5))
print("Varianza aprox: ",np.var(P5))
print("Desviacin estandar aprox: ",np.std(P5))
```

$\mathcal{DU(a,b)}$

$f(x = k)$ = \begin{cases}
\dfrac{1}{b - a + 1} & \text{si } k \in \{a, a+1, \dots, b\}, \\[0.8em]
0 & \text{en otro caso.}
\end{cases}


$F(x) =$
\begin{cases}
0 & \text{si } x < a, \\[0.5em]
\dfrac{\lfloor x \rfloor - a + 1}{b - a + 1} & \text{si } a \leq x \leq b, \\[0.5em]
1 & \text{si } x > b,
\end{cases}


$$E[x]=\frac{a+b}{2}$$
$$V[x]=\frac{(b-a+1)^2-1}{12}$$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def uniforme(n, a, b):
    uniform = []
    for _ in range(n):
        R = rd.uniform(a, b)
        x = a + (b-a)*R
        uniform.append(x)
    return uniform

A=uniforme(10000, 0, 1)
print('MEDIA:', np.mean(A))
print('VARIANZA:', np.var(A))
print('DESVIACIN ESTNDAR:', np.std(A))
plt.hist(A, bins=1000,color="pink")
plt.title("Uniforme f(x)")
plt.show()
```



*   **Bernoulli**


---


---







$\mathcal{B(p)} \qquad p \in (0,1)$

$f(x)$ =
\begin{cases}
p       & \text{si } x = 1, \\[0.5em]
1 - p   & \text{si } x = 0, \\[0.5em]
0       & \text{en otro caso.}

image.png


$F(x)$ =
\begin{cases}
0               & \text{si } x < 0, \\[0.5em]
1 - p           & \text{si } 0 \leq x < 1, \\[0.5em]
1               & \text{si } x \geq 1.

image.png
$$E[x]=p$$
$$V[x]=p(1-p)$$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def bernoullif(n, p):
    bernoulli = []#lista para almacenar los x
    for _ in range(n):#ciclo para generar los nmeros aleatorios
        R = rd.uniform(0, 1)
        if R <= p:#primer condicion sobre R
            x = 1
        else:#en otro caso
            x = 0
        bernoulli.append(x)#almacena el resultado de las condiciones
    return bernoulli

BERNOULLI=bernoullif(10, 0.5)
print('ME:', np.mean(BERNOULLI))
print('VAR:', np.var(BERNOULLI))
print('DES est:', np.std(BERNOULLI))
plt.hist(BERNOULLI, bins=100,color="pink")
plt.title("Bernoulli f(x)")
plt.show()
```



*   **Geomtrica**


---



---






$\mathcal{geom(p)} \qquad p \in (0,1)$

$f(x)=$
\begin{cases}
p(1-p)^x       & \text{si } x = 0, 1, \cdots  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}

$F(x)$ =
\begin{cases}
0 & \text{otro caso}, \\[0.5em]
1 - (1 - p)^{\lfloor x \rfloor +1} & \text{si } x \geq 0.

image.png
$$E[x]=\frac{1-p}{p}$$
$$V[x]=\frac{1-p}{p^2}$$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def geometricaf(n, p):
    geometric = []#lista vaca para almacenar los valores
    for _ in range(n):#ciclo para generar los nmeros aleatorios
        R = rd.uniform(0, 1)
        geometric.append(np.log(R)/np.log(1-p))#calcula y guarda los X
    return geometric

GEOME=geometricaf(1000, 0.5)
print('ME:', np.mean(GEOME))
print('VAR:', np.var(GEOME))
print('DESV EST:', np.std(GEOME))
plt.hist(GEOME, bins=100,color="pink")
plt.show()
```



*   **Binomial Negativa**


---



---





$\mathcal{negbin(s,p)} \qquad s>0 \qquad p \in (0,1)$

image.png

$f(x)=$
\begin{cases}
\binom{s + x - 1}{x} p^s (1-p)^x       & \text{si } x = 0,1,2,\cdot  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}



$F(x)=$
\begin{cases}
\sum_{i=0}^{x }\binom{s + i - 1}{i} p^s (1-p)^i       & \text{si } x 0  \\[0.5em]
0       & \text{en otro caso.}

$$E[x]=\frac{s(1-p)}{p}$$

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt

def binomialnegativa(n, s, p):
    binomial = [] #lista para almacenar los X
    for _ in range(n):#ciclo para generar los nmeros aleatorios
        R=np.random.geometric(p, s)
        binomial.append(np.sum(R))#clacula y guarda los X
    return binomial

Bin=binomialnegativa(10000, 10, 0.5)
plt.hist(Bin, bins=100,color="pink")
print('media:', np.mean(Bin))
print('var:', np.var(Bin))
print('Desviacion e:', np.std(Bin))
```



*   **POISSON**


---



---




$\mathcal{p(\lambda)} \qquad \lambda >0 \qquad x \in {0,1,2,\cdots}$

$f(x)=$
\begin{cases}
\frac{e^{-\lambda} \lambda^x}{x!}       & \text{si } x = 0,1,2,\cdot  \\[0.5em]
0       & \text{en otro caso.}
\end{cases}

image.png

$F(x)=$
\begin{cases}
0 & x < 0 \\[0.5em]
e^{-\lambda}\sum_{i=0}^{x }\frac{\lambda ^i}{i!} & \text{si } x \geq 0.
\end{cases}

image.png

$$E[x]=\lambda$$
$$V[x]=\lambda$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def poisson(lamda, n_muestras):
    lista = []
    for _ in range(n_muestras):#ciclo para generar los nmeros aleatorios
    #Parmetros
        a = math.exp(-lamda)
        b = 1.0
        i = 0#contador de evento
        while True:
            u = np.random.uniform(0, 1)
            b *= u
            if b < a:# Cuando b cae por debajo de e^(-), se detiene
                break
            i += 1 # Cada vez que b >= a, se incrementa el contador i
        lista.append(i)

    return lista

lista = poisson(4, 100)
print('MEDIA:', np.mean(lista))
print('VARIANZA:', np.var(lista))
print('DESVIACIN ESTNDAR:', np.std(lista))
# Graficar histograma
plt.hist(lista, bins=10,color="pink")
plt.show()
```
