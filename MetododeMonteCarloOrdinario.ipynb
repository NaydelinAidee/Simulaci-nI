Si tomamos una muestra de numeros pseudoaleatorios uniformemente distribuidos, digamos X1, ..., XN U∼ Unif(a, b), por la ley de los grandes numeros para N suficientemente grande tendremos que, con probabilidad 1:

A continuacion se presenta un codigo implementado en python donde se ilustra como funciona este metodo para aproximar la siguiente integral:

image.png

```python
import random
import numpy as np
from sympy import integrate, Symbol, sqrt, atan
import matplotlib.pyplot as plt

def Metodo_Monte_carlo (N,a,b):
  x = random.uniform(a,b)
  lista = []
  for n in range (N):
    lista.append(x)
  return lista

def g(x):
    return np.sqrt(np.arctan(x))

x = Metodo_Monte_carlo (1000,0,1)
promedio = np.mean(g(x))

print(promedio)
x = Symbol('x')
inte = integrate(sqrt(atan(x)), (x, 0, 1))
resul = inte.evalf()

# Cálculo del error
error = abs(resul - promedio)
# Impresión de resultados
print(f"Resultado de la integral: {resul:.5f}")
print(f"Error: {error:.5f}")
 # Gráfica de convergencia
plt.figure(figsize=(12, 6))
plt.plot(promedio, label='Promedio Monte Carlo', color='blue')
plt.axhline(y=resul, color='red', linestyle='--', label=f'Valor exacto: {resul:.5f}')
plt.xlabel('Número de Iteraciones')
plt.ylabel('Valor Promedio')
plt.title('Convergencia de Monte Carlo hacia el valor exacto de la integral')
plt.legend()
plt.grid(True)
plt.show()
```
