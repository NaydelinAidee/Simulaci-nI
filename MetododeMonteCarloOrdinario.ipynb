Primero mandamos a llamar a nuestras librerias que nos permiten generar numeros aleatorios y hacer operaciones matematicas como lo es el calculo de la raiz cuadrada.

```python
import random
import numpy as np
```

Ahora definimos nuestra funcion que implementa este metodo

1.-Genera x que sera un numero aleatorio entre 0 y 1

2.-Lo almacenamos en una lista  que va a tener N valores

3.-Retornamos dicha lista

```python
def Metodo_Monte_carlo (N,a,b):
  x = random.uniform(a,b)
  lista = []
  for n in range (N):
    lista.append(x)
  return lista
```

definimos nuestra funcion g(x) que es donde tenemos el ejemplo con el que vamos a trabajar,mandamos allamar a nuestra x y calculamos el promedio.

```python
def g(x):
    return np.sqrt(np.arctan(x))

x = Metodo_Monte_carlo (1000,0,1)
promedio = np.mean(g(x))

print(promedio)
```
