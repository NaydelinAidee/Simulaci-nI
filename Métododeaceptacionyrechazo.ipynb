# METODO DE ACEPTACION Y RECHAZO
**NAYDELIN AIDEE CARCAMO HERNANDEZ**

El método de aceptación y rechazo se usa  para generar números pseudoaleatorios a partir de una variable aleatoria. Lo utilizamos cuando la distribución de probabilidad de interés es difícil o imposible de simular directamente.se generan muestras de una distribución más fácil de simular y se aceptan o rechazan con una probabilidad.

**Los pasos parael metodo son:**


1.   Generar dos números aleatorios  $ (R1,R2)$
2.   Determinar  $ x = a + (b - a )R1 $
3.   Evaluar $f$ en $x$
4.   Si $R2 \leq \frac{f(x)}{M}$ entonces se acepta y en otro caso se rechaza



A continuacion mostraremos un ejemplo con la siguiente funcion triangular:

$f(x)= \left\{ \begin{array}{lcc} \frac{2}{(c-a)(b-a)}(x-a) & si & a \leq x
& \leq b  \\ \\ \frac{2}{(c-a)(b-c)}(x-c) & si & b < x < c \\ \\ \\\end{array} \right.$


importamos las librerias necesarias:


random ----> para los numero aleatorios

numpy -----> para nuestra lista

matplotlib-> para graficar

```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
```

Damos vaores para a,b,c y M

```python
a=1
b=3
c=5
M=2 * (c-a)
```

como tenemos una funcion a trozos enonces definimos 2 funciones:


```python
def funcion1(x):
    return (2 / (c-a)*(b-a)) * (x-a)
def funcion2(x):
    return (-2 / (c-a) * (c-b)) * (x-c)
```

Ahora declaramos una nueva funcion donde onemos las codiciones  y dependendo en que intervalo este es la funcion con la que se va a trabajar.

```python
def funcion(x):
    if x <= b and x>= a:
        return funcion1(x)
    else:
        return funcion2(x)
```

definimos la funcion donde realizaremos el metodo y aqui creamos una lista y un ciclo for y dentr de este creamos R1 Y R2 y obtenemos x,ademas incluimos un condicional y retornamos la lista.


```python
def metodoDeAceptacionYRechazo(n):
    lista = []
    for _ in range(n):
        R1 = rd.random()
        R2 = rd.random()
        x = a + (c - a) * R1
        if R2 <= funcion(x) / M:
            lista.append(x)
    return lista
```

Por ultimo mandamos a llamar nuestro metodo y visualizamos en la grafica.

```python
c = metodoDeAceptacionYRechazo(10000)
print(c)
plt.hist(c, bins=50)
plt.show()
```

Ahora calculamos la media y la varianza.

```python
print(np.var(c))
print(np.mean(c))
```
