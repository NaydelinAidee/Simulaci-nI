```python
import matplotlib.pyplot as plt
import numpy as np

# Define la función de la parábola (por ejemplo, y = x^2)
def parabola(x):
  return x**2

# Genera valores de x
x = np.linspace(-10, 10, 100)

# Calcula los valores correspondientes de y
y = parabola(x)

# Crea la gráfica
plt.plot(x, y)
plt.title('Gráfica de una parábola')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Muestra la gráfica
plt.show()
```
